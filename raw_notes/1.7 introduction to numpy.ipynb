### Заметка: Введение в NumPy (Introduction to NumPy)

**Основная концепция**
*   **NumPy** — это базовая библиотека для работы с массивами и вычислениями в Python, которая повсеместно используется в Data Science и Machine Learning.
*   Общепринятая конвенция (convention) при импорте — использовать сокращение **`np`**: `import numpy as np`.

**Создание одномерных массивов (Creating Arrays)**
NumPy предлагает несколько удобных функций для быстрой генерации массивов:
*   `np.zeros(size)` — создает массив (**array**), заполненный нулями.
*   `np.ones(size)` — создает массив, заполненный единицами.
*   `np.full(size, number)` — заполняет массив произвольным указанным числом (например, 2.5).
*   `np.array(list)` — конвертирует обычный список Python (**python list**) в массив NumPy.
*   `np.arange(start, end)` — генерирует диапазон чисел (по аналогии со стандартной функцией `range` в Python). Последнее число не включается в массив (**exclusive**).
*   `np.linspace(start, end, size)` — создает массив заданного размера, равномерно заполненный числами от начального до конечного значения (включительно).

**Многомерные массивы (Multi-dimensional arrays)**
*   В NumPy легко создавать двумерные (**two-dimensional**) и многомерные массивы. Для этого функциям вроде `zeros` нужно передать размерность в виде `(rows, columns)` — количество строк и столбцов.
*   Двумерный массив также можно создать, передав в `np.array()` список списков (**list of lists**).

**Индексация и доступ к элементам (Indexing)**
*   Как и в чистом Python, индексация в NumPy начинается с нуля.
*   Для доступа к элементам двумерного массива используются два индекса через запятую: `array[row_index, column_index]`.
*   **Slicing (срезы)**: чтобы получить всю строку, достаточно указать один индекс (например, `array`). Чтобы получить всю колонку, используется оператор двоеточия `[:]`, который означает «все строки» — например, `array[:, 1]` вернет колонку под индексом 1.
*   Любой выбранный элемент, строку или колонку можно перезаписать с помощью оператора присваивания (`=`).

**Генерация случайных чисел (Random generation)**
Пакет `np.random` позволяет генерировать массивы со случайными числами:
*   **Uniform distribution** (Равномерное распределение): генерация чисел от 0 до 1.
*   **Standard normal distribution** (Стандартное нормальное распределение): генерация чисел вокруг нуля.
*   **Random integers**: генерация целых чисел в заданном диапазоне (от `low` до `high`).
*   **Fixing the random seed**: Случайные числа в программировании на самом деле являются псевдослучайными (**pseudo-random**). Чтобы зафиксировать генератор и сделать результаты воспроизводимыми на разных компьютерах (**reproducible**), необходимо установить сид: `np.random.seed(number)`.

**Операции с массивами (Operations)**
1.  **Element-wise operations (Поэлементные операции)**
    *   В отличие от обычных списков Python, где потребовалось бы писать цикл `for`, в NumPy можно прибавлять, вычитать, умножать или делить числа прямо на весь массив (например, `array + 1` или `array * 100`).
    *   Операции можно объединять в цепочки (**chain**): `(array * 2) + 10`.
    *   Можно применять математические действия между двумя разными массивами одинакового размера — операция выполнится поэлементно.
2.  **Comparison operations (Операции сравнения)**
    *   Сравнения (например, `array >= 2` или `array1 > array2`) также выполняются поэлементно и возвращают массив булевых значений (`True` или `False`).
3.  **Boolean filtering (Фильтрация)**
    *   Если булево условие поместить в квадратные скобки к исходному массиву (например, `array[array >= 2]`), функция вернет только те элементы, для которых условие является истинным. 

**Агрегирующие операции (Summarizing operations)**
Вместо возвращения нового массива, эти операции вычисляют и возвращают одно число (статистику) по всему массиву:
*   `min()` — минимальное значение.
*   `max()` — максимальное значение.
*   `sum()` — сумма всех элементов.
*   `mean()` — среднее арифметическое.
*   `std()` — стандартное отклонение (**standard deviation**).
Эти операции работают как с одномерными, так и с многомерными массивами.
