### Заметка: Введение в Pandas (Introduction to Pandas)

**Основная концепция**
*   **Pandas** — это главная библиотека в Python для работы с табличными данными (**tabular data**). 
*   Стандартная конвенция при импорте библиотеки — использовать сокращение **`pd`**: `import pandas as pd`.
*   Под капотом Pandas использует массивы NumPy, поэтому многие принципы работы похожи. Получить чистый массив NumPy из структуры данных Pandas можно с помощью атрибута `.values`.

**Основные структуры данных**
1.  **DataFrame (Датафрейм)**: это главная абстракция библиотеки, представляющая собой обычную таблицу с данными.
2.  **Series (Серия)**: каждый отдельный столбец в датафрейме представляет собой объект Series. Датафрейм состоит из множества Series.

**Создание и первичный просмотр данных**
*   Датафрейм можно создать с помощью конструктора `pd.DataFrame()`.
*   Удобный способ создания — передать список словарей (**list of dictionaries**). В таком случае ключи словарей автоматически станут названиями колонок (**column names**).
*   Для конвертации датафрейма обратно в список словарей используется метод `.to_dict(orient='records')`.
*   Для быстрого просмотра загруженных данных (например, первых 5 строк) используется метод `.head()`.

**Работа со столбцами (Columns)**
*   Доступ к колонке возможен через точку (`df.make`) или через квадратные скобки (`df['make']`). Скобки обязательны, если в названии колонки есть пробелы или спецсимволы.
*   Для выбора сразу нескольких колонок нужно передать список их названий в двойных квадратных скобках: `df[['make', 'model', 'price']]`.
*   Создание или замена колонки происходит через присваивание: `df['id'] =`. Удаление колонки — с помощью оператора `del` (например, `del df['id']`).

**Индексы и доступ к строкам (Index & Rows)**
Каждая строка в датафрейме имеет свой идентификатор — **index** (по умолчанию это последовательность от 0 и далее, называемая `RangeIndex`).
*   **`loc` (Location)**: используется для доступа к строкам по их явному индексу (который может быть как числом, так и строкой, например, `a`, `b`, `c`).
*   **`iloc` (Positional Index)**: используется для доступа к строкам по их порядковому номеру (от 0 до конца), независимо от того, как выглядит явный индекс.
*   **Сброс индекса**: если индекс стал нестандартным, его можно вернуть к sequential (от 0 и далее) с помощью `df.reset_index(drop=True)`.

**Операции с данными (Operations & Filtering)**
*   **Element-wise operations (Поэлементные операции)**: работают так же, как в NumPy. Например, `df['engine_hp'] / 100` разделит все значения в колонке на 100. При этом пропущенные значения (`NaN` — Not a Number) игнорируются и остаются пропущенными.
*   **Boolean filtering (Фильтрация)**: для выбора определенных строк используется логическое условие внутри квадратных скобок: `df[df['year'] > 2015]`. Условия можно объединять логическим оператором И (`&`): `(df['make'] == 'nissan') & (df['year'] > 2015)`.

**Строковые операции (String operators)**
Для работы с текстовыми колонками (**categorical variables**) в Pandas есть специальный аксессор `.str`, позволяющий применять строковые методы сразу ко всей серии. 
*   Их можно объединять в цепочки (**chaining**), например: `df['vehicle_style'].str.lower().str.replace(' ', '_')` (перевод в нижний регистр и замена пробелов на нижнее подчеркивание).

**Агрегирующие операции и статистика (Summarizing operators)**
*   Как и в NumPy, доступны базовые функции: `.min()`, `.max()`, `.mean()` (среднее), `.std()`.
*   Метод **`.describe()`** — невероятно полезная функция, которая автоматически вычисляет количество, среднее, стандартное отклонение, минимум, максимум и процентили (**percentiles**) для всех числовых колонок датафрейма.
*   Функция **`.nunique()`** позволяет узнать количество уникальных значений в колонках (полезно для текстовых данных).
*   **Обработка пропусков**: комбинация `.isnull().sum()` позволяет быстро подсчитать, сколько пропущенных значений (**missing values**) находится в каждой колонке.

**Группировка данных (Grouping)**
*   В Pandas реализована функция **`groupby`**, аналогичная оператору `GROUP BY` в SQL.
*   Она позволяет сгруппировать данные по определенной категории, а затем вычислить статистику внутри каждой группы. Пример: `df.groupby('transmission_type').price.mean()` — найдет среднюю цену автомобиля для каждого типа трансмиссии (ручной и автоматической).
